### Conversation between Chatbots

```python
from openai import OpenAI
from IPython.display import Markdown, display, update_display
```

```python
openai = OpenAI(base_url='http://localhost:11434/v1/', api_key='ollama')
```

```python
Alex_model = 'nemotron-3-super:cloud'
Ben_model = 'gpt-oss:120b-cloud'
Clare_model = 'gemma4:31b-cloud'
```

```python
alex_system = """
You are Alex.
You are argumentative, skeptical, and sharp.
- Challenge assumptions
- Disagree often
- Be concise and slightly snarky
"""

ben_system = """
You are Ben.
You are cooperative and constructive.
- Build on ideas
- Find common ground
- Stay positive
"""

clare_system = """
You are Clare.
You are analytical and insightful.
- Reframe problems
- Add new perspectives
- Keep responses structured
"""
```

```python
def call_model(model, system_prompt, user_prompt):
    response = openai.chat.completions.create(
        model=model,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ]
    )
    
    return response.choices[0].message.content
```

```python
agents = [
    {
        "name": "Alex",
        "model": Alex_model,
        "system": alex_system
    },
    {
        "name": "Ben",
        "model": Ben_model,
        "system": ben_system
    },
    {
        "name": "Clare",
        "model": Clare_model,
        "system": clare_system
    }
]
```

```python
def build_user_prompt(name, conversation, last_speaker):
    """Build a prompt for the current agent based on conversation history."""
    return f"""{conversation}

{name}, please respond to the topic. Be mindful of what {last_speaker} just said."""
```
